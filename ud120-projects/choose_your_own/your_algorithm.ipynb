```python
def classify_adaboost(features_train, labels_train):
    
    ### your code goes here--should return a trained decision tree classifer
    from sklearn.ensemble import AdaBoostClassifier
    clf = AdaBoostClassifier()
    clf = clf.fit(features_train, labels_train)
    
    
    return clf
```

```python
def classify_SVC(features_train, labels_train):
    
    ### your code goes here--should return a trained decision tree classifer
    from sklearn.svm import SVC
    clf = SVC()
    clf = clf.fit(features_train, labels_train)
    
    
    return clf
```

```python
clf = classify_SVC(features_train, labels_train)
pred = clf.predict(features_test)
from sklearn.metrics import accuracy_score

acc = accuracy_score(labels_test, pred)
print acc





try:
    prettyPicture(clf, features_test, labels_test)
except NameError:
    pass
```

```python
def classify_KNN(features_train, labels_train):
    from sklearn.neighbors import KNeighborsClassifier
    clf = KNeighborsClassifier(n_neighbors = 4)
    clf = clf.fit(features_train, labels_train)
    return clf
```

```python
clf = classify_KNN(features_train, labels_train)
pred = clf.predict(features_test)
from sklearn.metrics import accuracy_score

acc = accuracy_score(labels_test, pred)
print acc





try:
    prettyPicture(clf, features_test, labels_test)
except NameError:
    pass
```
